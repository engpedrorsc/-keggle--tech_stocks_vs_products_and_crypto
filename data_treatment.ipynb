```python
import pandas as pd
import os
import glob
```

Importing dataset

```python
# Current working directory

print(f'Current Directiory: {os.getcwd()}')
```

```python
# Search, read and store in dictionary

datasets_dir = './original_datasets'
extensions = ['.csv']
df_dict = {}

for extension in extensions:
    for file in glob.glob(datasets_dir + f'**/**/*{extension}', recursive=True):
        file_path = file
        file_name = os.path.basename(file).split('.')[0]
        df_dict[file_name] = pd.read_csv(file_path)
```
